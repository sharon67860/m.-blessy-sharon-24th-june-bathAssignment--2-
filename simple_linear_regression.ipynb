```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.DataFrame({'years_Experience':[1,4,6,9,3,2,11,5,7],
                   'salary':[12,25,30,32,15,13,35,27,31]})
df
```

```python
X = df['years_Experience'].copy()#independent
Y = df['salary']#dependent
```

**Divide the data into trian and test**

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
len(X_train),len(Y_train)
```

```python
len(X_test),len(Y_test)
```

```python
training_data = pd.DataFrame()
training_data['X_train'] = X_train
training_data['Y_train'] = Y_train
training_data
```

```python
testing_data = pd.DataFrame()
testing_data['X_test'] =X_test
testing_data['Y_test'] = Y_test
testing_data
```
